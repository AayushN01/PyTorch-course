```python
a = torch.tensor(5.)
b = torch.tensor(6., requires_grad = True)
c = torch.tensor(2., requires_grad = True)
```

```python
d = a * b + c
d
```

```python
d.backward()
```

```python
print('dd/da =', a.grad)
print('dd/db =', b.grad)
print('dd/dc =', c.grad)
```
